```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/reference_html_as_sentences_df.csv')
```

```python
bad_netloc_aggs = [
    'witches.shca.ed.ac.uk','en.isabart.org','bechdeltest.com','npg.si.edu','www.guidetopharmacology.org',
    'letterboxd.com','www.discogs.com','vocab.getty.edu','www.isfdb.org', 'www.npg.org.uk',
    'art.nationalgalleries.org', 'www.tate.org.uk' ,'www.getty.edu', 'memory-beta.fandom.com', 'www.disease-ontology.org',
    'artgallery.yale.edu', 'www.imdb.com', 'muckrack.com', 'live.dbpedia.org', 'dbpedia.org'
]
df = df[~df.netloc_agg.isin(bad_netloc_aggs)].reset_index(drop=True)

df.info()
```

```python
df.reference_id.unique().shape[0]
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,20))

sns.countplot(
    data = df,
    y='netloc_agg',
    orient='h',
    order=df.netloc_agg.value_counts().index,
)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,5))

sns.countplot(
    data = df,
    y='reference_datatype',
    orient='h',
    order=df['reference_datatype'].value_counts().index,
)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,5))

sns.countplot(
    data = df,
    y='content-type',
    orient='h',
    order=df['content-type'].value_counts().index,
)
```
